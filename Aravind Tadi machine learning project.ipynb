```python
#polynomialregression with two variables

import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures

df=pd.read_csv("census.csv")
df

x=df[['Total Population','Median Age']]

y=df[['Average Household Size']]

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=5)

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train,y_train)

y_predict=model.predict(x_test)

from sklearn.metrics import mean_squared_error

mse=mean_squared_error(y_test,y_predict)

print(mse)

plt.plot(y_test,y_predict,'k',label='Linear Regression')

print(model.score(x_test,y_test))#Accuracy of Linear Regression

p1=PolynomialFeatures(degree=2)

x_train_poly1=p1.fit_transform(x_train)

x_test_poly1=p1.fit_transform(x_test)

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train_poly1,y_train)

print(model.score(x_test_poly1,y_test))#Accuracy of Polynomial Regression with degree 2

y_predict1=model.predict(x_test_poly1)

plt.plot(y_test,y_predict1,'ro',label='Polynomial Regression_2')

p2=PolynomialFeatures(degree=3)

x_train_poly2=p2.fit_transform(x_train)

x_test_poly2=p2.fit_transform(x_test)

from sklearn.linear_model import LinearRegression

model=LinearRegression()

model.fit(x_train_poly2,y_train)

print(model.score(x_test_poly2,y_test))#Accuracy of Polynomial Regression with degree 3

y_predict2=model.predict(x_test_poly2)

plt.plot(y_test,y_predict2,'bo',label='Polynomial Regression_3')

plt.xlabel("Total Population','Median Age'")

plt.ylabel("'Average Household Size'")

plt.title("Linear vs Polynomial Regression")

 


```
